これで属性と名前，年齢，誕生日，星座などなどが全部取れたので，要素ごとにリストを作ってそこに突っ込むことにする


今回の使用用途的には絶対いらない情報（星座，誕生日などなど）もせっかくなので一度Dataframeにしてしまおう

````python
attribute = []
name = []
age =[]
birthday = []
constellation = []
blood_type = []
height = []
weight = []
bust = []
waist = []
hip = []
handedness = []
birthplace = []
hobby = []
character_voice = []


for tr in table.find_all("tr"):
    idol_td = [elem.text for elem in tr.find_all("td")]
    attribute.append(idol_td[0])
    name.append(idol_td[1])
    age.append(idol_td[2])
    birthday.append(idol_td[3])
    constellation.append(idol_td[4])
    blood_type.append(idol_td[5])
    height.append(idol_td[6])
    weight.append(idol_td[7])
    bust.append(idol_td[8])
    waist.append(idol_td[9])
    hip.append(idol_td[10])
    handedness.append(idol_td[11])
    birthplace.append(idol_td[12])
    hobby.append(idol_td[13])
    character_voice.append(idol_td[14])
````

````python
import pandas # 思考停止でpandas
````

````python
# 最初から辞書にしとけばよかった感
cg_df = pandas.DataFrame({
    "attribute": attribute,
    "name" : name,
    "age" : age,
    "birthday" : birthday,
    "constellation" : constellation,
    "blood_type" : blood_type,
    "height" : height,
    "weight" : weight,
    "bust" : bust,
    "waist" : waist,
    "hip" : hip,
    "handedness" : handedness,
    "birthplace" : birthplace,
    "hobby" : hobby,
    "character_voice" : character_voice
})
````
